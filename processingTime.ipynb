```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

file_path = 'ProcessingTime/processing_time.txt'
separator = ';'
column_names = ['conversion', 'reconstruction', 'ipfs_upload', 'signature', 'bc', 'verify_signature', 'ipfs_download']

df = pd.read_csv(file_path, sep=separator, names=column_names)
print(df)
```

```python
# Creating a figure and axis
fig, ax = plt.subplots()

m1_conversion = df['conversion'].mean()
st1_conversion = df['conversion'].std()

m1_reconstruction = df['reconstruction'].mean()
st1_reconstruction = df['reconstruction'].std()

print("Conversion - Mean:", m1_conversion, "Std:", st1_conversion)
print("Reconstruction - Mean:", m1_reconstruction, "Std:", st1_reconstruction)

ax.grid(True, linestyle="dotted", color='grey')

# Creating boxplots
bp = ax.boxplot([df['conversion'], df['reconstruction']], showmeans=True)

# Annotate the boxplots
for i, line in enumerate(bp['medians']):
    x, y = line.get_xydata()[1]
    text = ' μ={:.2f}\n σ={:.2f}'.format(m1_conversion if i == 0 else m1_reconstruction,
                                         st1_conversion if i == 0 else st1_reconstruction)
    ax.annotate(text, xy=(x, y))

# Set the xticks
xtick_labels = ['Parameter Conversion', 'Parameter Reconstruction']  # Labels for each boxplot
ax.set_xticks([1, 2])  # Position of the xticks
ax.set_xticklabels(xtick_labels)  # Assigning labels to the xticks

# Adding labels and title
plt.xlabel('Mechanism')
plt.ylabel('Time (ms)')
plt.title('SECC - Parameter')

plt.savefig('fl_parameter_conversion_processing_time.png',bbox_inches='tight',  dpi=700)	#save image
plt.tight_layout()
plt.show()
```
